## Linear Regression with One Feature


- Linear model is sum weighted predicted data to target values
- Y = mx+c
- Y = target value/output
- m = slope
- x = input value
- c = Coefficient/Intercept

### Slope Formula
- m = (x-xmean)*(y-ymean)/(x-xmean)^2

### Coefficient or intercept formula
- c = ymean-(m*xmean) 


## ML steps:
- 1.Get or load the data
- 2.Preprocessing the data
- 3.Define input and output
- 4.Apply the model or algorithm
- 5.Pass the data to training and testing
- 6.Calcute the score or accuracy score


```python
## get the libraries
import pandas as pd
import numpy as np
```

```python
# 1.Get the data
df = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/Salary_Data.csv")
```

```python
df.head()
```

```python
# 2.Preprocessing the data
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
## 3. Define input and output
x=df[['YearsExperience']]
y=df['Salary']
```

```python
## 4. Apply the model or algorithm
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
## 5.Test the data
model.predict([[7.1],[10]])
```

```python
df
```

```python
# 6.Identify the score
model.score(x,y)*100
```

```python
# Visualize the relationship of input and output
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],label = 'actual data')
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],label = 'actual data')
plt.plot(x,y_pred,color = 'r',label = 'predicted values')
plt.title('Years of Experience vs Salary')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.legend()
plt.show()
```
